##**Создаем выборки**

```python
COL_MONTH = df2.columns.get_loc('месяц_desc')
COL_RATING = df2.columns.get_loc('рейтинг')
COL_TOTAL = df2.columns.get_loc('всего')
COL_ANSWER = df2.columns.get_loc('есть ответ')
COL_COMPLETE = df2.columns.get_loc('выполнен')
COL_CLOSED = df2.columns.get_loc('закрыт')
COL_NPS = df2.columns.get_loc('nps')
COL_EMAIL = df2.columns.get_loc('email')
COL_TELEPHONY = df2.columns.get_loc('telephony')
COL_CHAT = df2.columns.get_loc('usedesk_chat')
COL_PROBLEM = df2.columns.get_loc('problem')
COL_MEDIUM = df2.columns.get_loc('medium')
COL_CNT = df2.columns.get_loc('правила_cnt')
COL_DIST = df2.columns.get_loc('правила_dist')
COL_LICENS = df2.columns.get_loc('licens')
COL_BZ = df2.columns.get_loc('бз')
COL_ST = df2.columns.get_loc('бз_ст')
```

В связи с отсутствием категориальных данных (рейтинг мы уже привели к значеями 0-1) считаю разумным остальные данные привести к линейному min/max нормированию

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
def normalize_data(arg):
  norm_data = scaler.fit_transform(arg.reshape(-1,1))
  return norm_data
```

```python
def extract_row_data(row):
    # Извлечение и преобразование данных
    month = row[COL_MONTH]
    total = row[COL_TOTAL]
    answer = row[COL_ANSWER]
    complete = row[COL_COMPLETE]
    closed = row[COL_CLOSED]
    nps = row[COL_NPS]
    email = row[COL_EMAIL]
    telephony = row[COL_TELEPHONY]
    chat = row[COL_CHAT]
    problem = row[COL_PROBLEM]
    medium = row[COL_MEDIUM]
    cnt = row[COL_CNT]
    dist = row[COL_DIST]
    licens = row[COL_LICENS]
    bz = row[COL_BZ]
    st = row[COL_ST]
    rating = row[COL_RATING]

    # Объединение всех входных данных в один общий вектор
    x_data = np.hstack([month,
                        answer,
                        complete,
                        closed,
                        nps,
                        email,
                        telephony,
                        chat,
                        problem,
                        medium,
                        cnt,
                        dist,
                        licens,
                        bz,
                        st])

    # Возврат входных данных и выходных
    return x_data, rating
```

```python
def construct_train_data(row_list):
    x_data = []
    y_data = []

    for row in row_list:
        x, y = extract_row_data(row)
        x_data.append(x)
        y_data.append(y)

    return np.array(x_data), np.array(y_data)
```

```python
df_scaled = df2.copy()
for column in df_scaled.columns:
    df_scaled[column] = normalize_data(df_scaled[column].values)
x_data, y_data = construct_train_data(df_scaled.values)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, shuffle = True, random_state=4)
```

#**Собираем тестовые модели нейронной сети**

```python
# Последовательная модель НС
from tensorflow.keras.models import Sequential

# Основные слои
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization

# Утилиты для to_categorical()
from tensorflow.keras import utils

# Алгоритмы оптимизации для обучения модели
from tensorflow.keras.optimizers import Adam
```

```python
from sklearn.metrics import accuracy_score

# Функция оценки точности нейросети
def eval_net_acc(model, x_train, y_train, n = 10):
    pred = model.predict(x_train)             # Предсказание на тренировочной выборке
    pred = np.round(pred).astype(int)         # Округление предсказаний до ближайшего целого числа

    print('Точность:', accuracy_score(y_train, pred), '\n')

    # Печатаем первые n примеров результата
    for i in range(min(n, len(y_train))):
        print('Реальное значение:', y_train[i], ' Предсказанное значение:', pred[i])
```

```python
def plot_history(history):
    plt.plot(history.history['accuracy'],
            label='Точность на обучающем наборе')
    plt.plot(history.history['val_accuracy'],
            label='Точность на валидационном наборе')
    plt.xlabel('Эпоха обучения')
    plt.ylabel('Точность')
    plt.legend()
    plt.show()
```

#**Oversampling и Undersampling данных**

Для начала проверим, что у нас по распределению на классы - возможен перевес в сторону класса один, поэтому сеть предпочитает склоняться именно к нему

```python
print('Обучающая выборка параметров', x_train.shape)
print('Обучающая выборка меток классов', y_train.shape)
print()
print('Тестовая выборка параметров', x_test.shape)
print('Тестовая выборка меток классов', y_test.shape)
orig = np.unique(y_data, return_counts=True)
train = np.unique(y_train, return_counts=True)
test = np.unique(y_test, return_counts=True)

for i in range(orig[0].shape[0]):
    print(f'{int(orig[0][i])} --> {orig[1][i]} |    {train[1][i]} | {test[1][i]}  |')

class_labels = ['0', '1']

fig, ax = plt.subplots(figsize=(5, 6))

ax.bar(class_labels, orig[1], label='test', color='m')
ax.bar(class_labels, train[1], label='Train', color='b', alpha=0.7)


ax.set_title('Распределение классов')
ax.set_xlabel('Номер класса')
ax.set_ylabel('Количество')
ax.legend()
plt.show()
```

Как видно, 1 класс сильно доминирует над 0, что может являться причиной неверной работы нейронной сети.

```python
def extract_row_data_02(row):
    # Извлечение и преобразование данных
    month = row[COL_MONTH]
    total = row[COL_TOTAL]
    answer = row[COL_ANSWER]
    complete = row[COL_COMPLETE]
    closed = row[COL_CLOSED]
    nps = row[COL_NPS]
    email = row[COL_EMAIL]
    telephony = row[COL_TELEPHONY]
    api = row[COL_API]
    usdchat = row[COL_USDCHAT]
    vk = row[COL_VK]
    tg = row[COL_TG]
    whatsapp = row[COL_WHATSAPP]
    fbig = row[COL_FBIG]
    chat = row[COL_CHAT]
    ig = row[COL_IG]
    fb = row[COL_FB]
    ptp = row[COL_PTP]
    twment = row[COL_TWMENT]
    c2d = row[COL_C2D]
    slack = row[COL_SLACK]
    twdir = row[COL_TWDIR]
    question = row[COL_QUESTION]
    task = row[COL_TASK]
    problem = row[COL_PROBLEM]
    incident = row[COL_INCIDENT]
    low = row[COL_LOW]
    medium = row[COL_MEDIUM]
    urgent = row[COL_URGENT]
    extreme = row[COL_EXTREME]
    cnt = row[COL_CNT]
    dist = row[COL_DIST]
    licens = row[COL_LICENS]
    bz = row[COL_BZ]
    st = row[COL_ST]
    rating = row[COL_RATING]

    # Объединение всех входных данных в один общий вектор
    x_data = np.hstack([month,
                        answer,
                        complete,
                        closed,
                        nps,
                        email,
                        telephony,
                        api,
                        usdchat,
                        vk,
                        tg,
                        whatsapp,
                        fbig,
                        chat,
                        ig,
                        fb,
                        ptp,
                        twment,
                        c2d,
                        slack,
                        twdir,
                        question,
                        task,
                        problem,
                        incident,
                        low,
                        medium,
                        urgent,
                        extreme,
                        cnt,
                        dist,
                        licens,
                        bz,
                        st])

    # Возврат входных данных и выходных
    return x_data, rating
```

```python
def construct_train_data_02(row_list):
    x_data_02 = []
    y_data_02 = []

    for row in row_list:
        x, y = extract_row_data_02(row)
        x_data_02.append(x)
        y_data_02.append(y)

    return np.array(x_data_02), np.array(y_data_02)
```
