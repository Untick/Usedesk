**Создаем выборки**

```python
COL_MONTH = df2.columns.get_loc('месяц_desc')
COL_RATING = df2.columns.get_loc('рейтинг')
COL_ANSWER = df2.columns.get_loc('есть ответ')
COL_COMPLETE = df2.columns.get_loc('выполнен')
COL_CLOSED = df2.columns.get_loc('закрыт')
COL_NPS = df2.columns.get_loc('nps')
COL_EMAIL = df2.columns.get_loc('email')
COL_TELEPHONY = df2.columns.get_loc('telephony')
COL_CHAT = df2.columns.get_loc('usedesk_chat')
COL_PROBLEM = df2.columns.get_loc('problem')
COL_MEDIUM = df2.columns.get_loc('medium')
COL_CNT = df2.columns.get_loc('правила_cnt')
COL_DIST = df2.columns.get_loc('правила_dist')
COL_LICENS = df2.columns.get_loc('licens')
COL_BZ = df2.columns.get_loc('бз')
COL_ST = df2.columns.get_loc('бз_ст')
```

В связи с отсутствием категориальных данных (рейтинг мы уже привели к значеями 0-1) считаю разумным остальные данные привести к линейному min/max нормированию

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
def normalize_data(arg):
  norm_data = scaler.fit_transform(arg.reshape(-1,1))
  return norm_data
```

```python
def extract_row_data(row):

    # Извлечение и преобразование данных
    month = normalize_data(row[COL_MONTH])
    answer = normalize_data(row[COL_ANSWER])
    complete = normalize_data(row[COL_COMPLETE])
    closed = normalize_data(row[COL_CLOSED])
    nps = normalize_data(row[COL_NPS])
    email = normalize_data(row[COL_EMAIL])
    telephony = normalize_data(row[COL_TELEPHONY])
    chat = normalize_data(row[COL_CHAT])
    problem = normalize_data(row[COL_PROBLEM])
    medium = normalize_data(row[COL_MEDIUM])
    cnt = normalize_data(row[COL_CNT])
    dist = normalize_data(row[COL_DIST])
    licens = normalize_data(row[COL_LICENS])
    bz = normalize_data(row[COL_BZ])
    st = normalize_data(row[COL_ST])
    rating = row[COL_RATING]

    # Объединение всех входных данных в один общий вектор
    x_data = np.hstack([month,
                        answer,
                        complete,
                        closed,
                        nps,
                        email,
                        telephony,
                        chat,
                        problem,
                        medium,
                        cnt,
                        dist,
                        licens,
                        bz,
                        st])

    # Возврат входных данных и выходных
    return x_data, rating


# Создание общей выборки
def construct_train_data(row_list):
    x_data = []
    y_data = []

    for row in row_list:
        x, y = extract_row_data(row)
        x_data.append(x)
        y_data.append(y)

    return np.array(x_data), np.array(y_data)

# Формирование выборки из загруженного набора данных
x_data, y_data = construct_train_data(df2.values)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=4)
```

**Собираем модельнейронной сети**

```python
# Последовательная модель НС
from tensorflow.keras.models import Sequential

# Основные слои
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization

# Утилиты для to_categorical()
from tensorflow.keras import utils

# Алгоритмы оптимизации для обучения модели
from tensorflow.keras.optimizers import Adam
```

```python
from sklearn.metrics import mean_absolute_error
# Функция оценки точности нейросети
def eval_net(model, x_train, y_train, y_scaler = None, n = 10, limit = 1.):

    pred = model.predict(x_train)             # Предсказание на тренировочной выборке
    y_train_norm = y_train.reshape(-1, 1)

    print('Средняя абсолютная ошибка:', mean_absolute_error(pred, y_train), '\n')

    for i in range(n):
        print('Реальное значение: {:6.2f}  Предсказанное значение: {:6.2f}  Разница: {:6.2f}'.format(y_train_norm[i, 0],
                                                                                                    pred[i, 0],
                                                                                                    abs(y_train_norm[i, 0] - pred[i, 0])))
    # Построение картинки разброса предсказанных и верных значений зарплаты
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.scatter(y_train, pred)                 # Отрисовка точечного графика
    ax.set_xlim(0, limit)                     # Пределы по x, y
    ax.set_ylim(0, limit)
    ax.plot(plt.xlim(), plt.ylim(), 'r')      # Отрисовка диагональной линии
    plt.xlabel('Правильные значения')
    plt.ylabel('Предсказания')
    plt.grid()
    plt.show()
```

```python
def plot_history(history):
    plt.plot(history.history['mae'],
            label='Средняя абсолютная ошибка на обучающем наборе')
    plt.plot(history.history['val_mae'],
            label='Средняя абсолютная ошибка на проверочном наборе')
    plt.xlabel('Эпоха обучения')
    plt.ylabel('Средняя абсолютная ошибка')
    plt.legend()
    plt.show()
```

```python
model = Sequential()

model.add(Dense(15, activation='relu', input_shape=(x_train_01.shape[1], )))

model.add(Dense(150, activation='tanh'))
model.add(BatchNormalization())

model.add(Dense(10, activation='relu'))

model.add(Dense(1, activation= 'linear'))

model.compile(optimizer=Adam(learning_rate=1e-3), loss='mse', metrics=['mae'])

history =  model.fit(x_train_01,
                     y_train,
                     batch_size=60,
                     epochs=30,
                     validation_split=0.15,
                     verbose=1)
plot_history(history)
```

```python
eval_net(model, x_test_01, y_test)
```
